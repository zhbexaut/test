```python
day_1_users = df[df['date'] == '2023-11-01']['user_id'].unique()
day_2_users = df[df['date'] == '2023-11-02']['user_id'].unique()
retained = set(day_1_users) & set(day_2_users)
retention = len(retained) / len(day_1_users) * 100
retention
```

```python
users_total = df['user_id'].nunique()

users_with_views = df[df['view_adverts'] > 0]['user_id'].nunique()

conversion = users_with_views / users_total * 100
conversion
```

```python
avg_views = df.groupby('user_id')['view_adverts'].sum().mean()
avg_views
```

```python
df_ab = pd.read_excel(
    'Данные для тестового задания.xlsx',
    sheet_name='Данные АБ тестов'
)
df_ab.info()
df_ab.head()
```

```python
from scipy import stats

exp = df_ab[df_ab['experiment_num'] == 1]

a = exp[exp['experiment_group'] == 'control']['revenue']
b = exp[exp['experiment_group'] == 'test']['revenue']

stats.ttest_ind(a, b, equal_var=False)
```
